# soal & diketahui

```python
import pandas as pd
import seaborn as sb
```

```python
titanic_df = sb.load_dataset('titanic')
titanic_df.head()
```

Tugas :

1.   Dapatkan ukuran/dimensi dari dataset diatas, cek jumlah data kosong per kolom, lalu drop kolom-kolom tersebut. Selanjutnya cek data statistiknya, dan cek lagi ukuran/dimensi yang baru.
2.   Buat plot frekuensi untuk kolom-kolom ["survived", "pclass", "sex", "who"].



# jawaban soal no. 1

```python
#mencari ukuran/dimensi dataset
jumlah_baris, jumlah_kolom = titanic_df.shape

print("Jumlah baris : ", jumlah_baris)
print("Jumlah_kolom : ", jumlah_kolom)
```

```python
#mengecek jumlah data kosong per kolom
titanic_df.isna().sum()
```

```python
#drop kolom yang ada data kosongnya(NaN)
new_titanic_df = titanic_df.drop(labels=['age', 'embarked', 'deck', 'embark_town'], axis=1)

new_titanic_df
```

```python
#cek data statistik dari latest dataset
new_titanic_df.describe()
```

```python
#dimensi dari latest dataset
jumlah_baris, jumlah_kolom = new_titanic_df.shape

print("Jumlah baris : ", jumlah_baris)
print("Jumlah_kolom : ", jumlah_kolom)
```

# jawaban soal no. 2

```python
#plot frekuensi dari kolom survived, pclass, sex, who.
import matplotlib.pyplot as plt

kolom_survived = 'survived'

sb.countplot(x=kolom_survived, data=new_titanic_df)
plt.title(f'Plot Frekuensi {kolom_survived}')
plt.xlabel(kolom_survived)
plt.ylabel('Frekuensi')
plt.show()

kolom_pclass = 'pclass'

sb.countplot(x=kolom_pclass, data=new_titanic_df)
plt.title(f'Plot Frekuensi {kolom_pclass}')
plt.xlabel(kolom_pclass)
plt.ylabel('Frekuensi')
plt.show()

kolom_who = 'who'

sb.countplot(x=kolom_who, data=new_titanic_df)
plt.title(f'Plot Frekuensi {kolom_who}')
plt.xlabel(kolom_who)
plt.ylabel('Frekuensi')
plt.show()

kolom_sex = 'sex'

sb.countplot(x=kolom_sex, data=new_titanic_df)
plt.title(f'Plot Frekuensi {kolom_sex}')
plt.xlabel(kolom_sex)
plt.ylabel('Frekuensi')
plt.show()
```
